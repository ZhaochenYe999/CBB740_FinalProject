```python
import pandas as pd
triage = pd.read_csv('triage.csv')
triage.head()
```

Note: I didn't separate entries in cases where there were multiple complaints because CountVectorizer already splits on whitespace and punctuation by default.

```python
# Preprocessing
triage["chiefcomplaint"] = (
    triage["chiefcomplaint"]
        .str.lower()
        .str.replace(",", " ")
        .str.replace("!", " ")
        .str.replace("?", " ")
        .str.strip()
)
triage["chiefcomplaint"] = triage["chiefcomplaint"].str.split().str.join(" ")
```

```python
# Feature extraction
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(triage['chiefcomplaint'])
print(X.toarray())
print(vectorizer.get_feature_names_out())
```
